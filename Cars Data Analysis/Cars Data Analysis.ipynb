<h3>1) Instruction ( For Data Cleaning ) - Find all Null Value in the dataset. If there is any null value in any column, then fill it with the mean of that column.</h3>

```python
car.isnull().sum()
```

```python
car.head(2)
```

```python
car.info()
```

```python
car['MSRP'] = car['MSRP'].replace(r'[\$,]', '', regex=True).astype(float)

car['Invoice'] = car['Invoice'].replace(r'[\$,]', '', regex=True).str.strip().astype(float)
```

```python
car.head(1)
```

```python
car.isnull().sum()
```

```python
cols = ['Make','Model','Type','Origin','DriveTrain']
for col in cols:
    car[col] = car[col].fillna(car[col].mode()[0])


num_cols = ['MSRP','Invoice','EngineSize','Cylinders','Horsepower','MPG_City','MPG_Highway','Weight','Wheelbase','Length']

car[num_cols] = car[num_cols].fillna(car[num_cols].mean())
```

```python
car.isnull().sum()
```
